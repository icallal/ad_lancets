```python
import pandas as pd

lancets = pd.read_parquet('../../../../randy/rfb/tidy_data/UKBiobank/dementia/lancet2024/lancet2024_preprocessed.parquet', engine = 'fastparquet')
```

```python
# ['RID', 'GDTOTAL', 'MH14BALCH', 'MH16BSMOK', 'AGE', 'PTGENDER',
#        'PTEDUCAT', 'HMHYPERT', 'DXAD', 'CLINICAL_LDL_C', 'bmi', 'apoe_2/2',
#        'apoe_2/3', 'apoe_2/4', 'apoe_3/3', 'apoe_3/4', 'apoe_4/4']

columns = ['eid', 'depression', 'alcohol_consumption', '20161-0.0', 'hypertension', '30780-0.0']
lancets = lancets[columns]
```

```python
covs = pd.read_parquet('doubleML_covariates.parquet', engine = 'fastparquet')
covs
```

```python
covs = covs[['IID', '31-0.0', 'curr_age', 'e3/e3', 'e2/e3', 'e3/e4', 'e2/e4', 'e4/e4', 'e2/e2', 'education_years', 'bmi', 'groups']]
```

```python
df = lancets.merge(covs, left_on = 'eid', right_on = 'IID', how = 'inner')
```

```python
df = df.drop(columns='IID')
```

```python
df['20161-0.0'] = df['20161-0.0'].fillna(0)
df.dropna(subset = ['education_years', '30780-0.0', 'bmi'], inplace=True)
```

```python
df.to_parquet('validation_data.parquet', index = False, engine = 'fastparquet')
```
